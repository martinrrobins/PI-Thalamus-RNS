## Boxplot

```python
df_f1       = pd.DataFrame({'iESPnet': f1_iespnet,       'DSF-iESPnet': f1_lopo})
df_accuracy = pd.DataFrame({'iESPnet': accuracy_iespnet, 'DSF-iESPnet': accuracy_lopo})
df_mae      = pd.DataFrame({'iESPnet': MAE_time_iespnet, 'DSF-iESPnet': MAE_time_lopo})

font_path        = '../03-Letra-plot/Montserrat-Regular.ttf'
montserrat       = fm.FontProperties(fname = font_path)

indices_thalamus = [2, 4, 9, 18, 25, 28]
thalamus_data_f1       = df_f1.loc[indices_thalamus]  
thalamus_data_accuracy = df_accuracy.loc[indices_thalamus]  
thalamus_data_mae      = df_mae.loc[indices_thalamus]  

fig, axs = plt.subplots(1, 3, figsize=(15, 6), facecolor='#000000')
fig.suptitle('LOPO 20 epocas', fontproperties=montserrat, fontsize=20, color='#FFFFFF')

for ax in axs.flat:  
    
    ax.spines['bottom'].set_color('#FFFFFF')
    ax.spines['top'   ].set_color('#FFFFFF')
    ax.spines['left'  ].set_color('#FFFFFF')
    ax.spines['right' ].set_color('#FFFFFF')
    
    ax.tick_params(axis='x', colors='#FFFFFF')
    ax.tick_params(axis='y', colors='#FFFFFF')
    

# boxplot F1
sns.boxplot(
            data         = df_f1, 
            palette      = ["#AF42AE", "#000000"], 
            ax           = axs[0],
            medianprops  = dict(color     = 'white'),  
            boxprops     = dict(edgecolor = 'white'),  
            whiskerprops = dict(color     = '#000000'), 
            capprops     = dict(color     = '#000000')
           )

sns.stripplot(data = df_f1.drop(indices_thalamus), size = 9, jitter = False, linewidth = 0.8, alpha = 0.6, palette = ['#AF42AE', '#000000'], edgecolor = '#FFFFFF', ax = axs[0])
sns.stripplot(data = thalamus_data_f1, size = 9, jitter = False, linewidth = 0.8, alpha = 1, palette = ['#2191FB', '#2191FB'], edgecolor = '#FFFFFF', ax = axs[0])

#leyenda
talamo_legend = Line2D([0], [0], marker = 'o', color = 'w', markerfacecolor = '#2191FB', markersize = 10, label = "Tálamo")
legend        = axs[0].legend(handles = [talamo_legend], loc = 'lower center', prop = {'size': 8})
for text in legend.get_texts():
    text.set_fontproperties(montserrat)

for i in indices_thalamus:
    axs[0].plot([0, 1], [f1_iespnet[i], f1_lopo[i]], linestyle='--', linewidth=0.8, alpha=1, color='#536B78')

axs[0].set_title('F 1', fontproperties = montserrat, fontsize = 16, color = '#FFFFFF')
axs[0].set_ylim([0, 1])

# boxplot Accuracy 
sns.boxplot(
            data         = df_accuracy, 
            palette      = ["#AF42AE", "#000000"], 
            ax           = axs[1],
            medianprops  = dict(color     = 'white'),  
            boxprops     = dict(edgecolor = 'white'),  
            whiskerprops = dict(color     = '#000000'), 
            capprops     = dict(color     = '#000000')
           )

sns.stripplot(data = df_accuracy.drop(indices_thalamus), size = 9, jitter = False, linewidth = 0.8, alpha = 0.6, palette = ['#AF42AE', '#000000'], edgecolor = '#FFFFFF', ax = axs[1])
sns.stripplot(data = thalamus_data_accuracy, size = 9, jitter = False, linewidth = 0.8, alpha = 1, palette = ['#2191FB', '#2191FB'], edgecolor = '#FFFFFF', ax = axs[1])

#leyenda
talamo_legend = Line2D([0], [0], marker = 'o', color = 'w', markerfacecolor = '#2191FB', markersize = 10, label = "Tálamo")
legend        = axs[1].legend(handles = [talamo_legend], loc = 'lower center', prop = {'size': 8})
for text in legend.get_texts():
    text.set_fontproperties(montserrat)

for i in indices_thalamus:
    axs[1].plot([0, 1], [accuracy_iespnet[i], accuracy_lopo[i]], linestyle='--', linewidth=0.8, alpha=1, color='#536B78')

axs[1].set_title('Accuracy balanceada', fontproperties = montserrat, fontsize = 16, color = '#FFFFFF')
axs[1].set_ylim([0, 1])

# boxplot MAE 
sns.boxplot(
            data         = df_mae, 
            palette      = ["#AF42AE", "#000000"], 
            ax           = axs[2],
            medianprops  = dict(color     = 'white'),  
            boxprops     = dict(edgecolor = 'white'),  
            whiskerprops = dict(color     = '#000000'), 
            capprops     = dict(color     = '#000000')
           )

sns.stripplot(data = df_mae.drop(indices_thalamus), size = 9, jitter = False, linewidth = 0.8, alpha = 0.6, palette = ['#AF42AE', '#000000'], edgecolor = '#FFFFFF', ax = axs[2])
sns.stripplot(data = thalamus_data_mae, size = 9, jitter = False, linewidth = 0.8, alpha = 1, palette = ['#2191FB', '#2191FB'], edgecolor = '#FFFFFF', ax = axs[2])

#leyenda
talamo_legend = Line2D([0], [0], marker = 'o', color = 'w', markerfacecolor = '#2191FB', markersize = 10, label = "Tálamo")
legend        = axs[2].legend(handles = [talamo_legend], loc = 'lower center', prop = {'size': 8})
for text in legend.get_texts():
    text.set_fontproperties(montserrat)

for i in indices_thalamus:
    axs[2].plot([0, 1], [MAE_time_iespnet[i], MAE_time_lopo[i]], linestyle='--', linewidth=0.8, alpha=1, color='#536B78')

axs[2].set_title('MAE time', fontproperties = montserrat, fontsize = 16, color = '#FFFFFF')
axs[2].set_ylim([0, 25])

for ax in axs:
    ax.set_xticklabels(ax.get_xticklabels(), fontproperties=montserrat, fontsize=14)
    ax.set_yticklabels(ax.get_yticklabels(), fontproperties=montserrat, fontsize=14)

plt.show()
```
